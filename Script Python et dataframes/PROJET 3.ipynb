<B> <span> 1ère question : CALCUL DE LA POPULATION MONDIALE  </span>  </B>

<B> <span> 2ème question : IDENTIFIER LES REDONDANCES  </span>  </B>

disponibilité intérieure = Production + importations - exportations + évolution des stocks

Mais je cherche aussi les emplois, je vais vérifier que l'on peut aussi dire : disponibilité intérieure = aliments pour animaux + semences + pertes + traitements + autres utilisations + nourriture

```python
# J'importe mon nouveau DataFrame complet du bilan alimentaire sans l'agrégation "Chine"

df_bilan = pd.read_csv("FAOSTAT_BILAN_ALIMENTAIRE_COMPLET.csv")
df_bilan.columns = ['Code Domaine','Domaine','Code Pays','Pays','Code Élément','Élément','Code Produit','Produit','Code Année','Année','Unité','Valeur','Symbole','Description du Symbole']
```

```python
# Je ne garde que les colonnes qui m'intéressent(Je ne veux pas le Code Domaine,Domaine )

df_bilan = df_bilan[['Code Pays','Pays','Code Élément','Élément','Code Produit','Produit','Code Année','Année','Unité','Valeur','Symbole','Description du Symbole']]
```

```python
# Je vérifie la formule "disponibilité intérieure = Production + importations - exportations + évolution des stocks"
# avec la France et le blé

df_bilan.loc[(df_bilan['Pays'].str.contains('France')) & (df_bilan['Produit'].str.contains('Blé'))]
```

<B> Calculez (pour chaque pays et chaque produit) la disponibilité alimentaire en kcal </B>

```python
# je veux créer un DataFrame ne contenant que la disponibilité alimentaire (Kcal/personne/jour) pour chaque produit

df_food_supply = df_bilan[ df_bilan['Code Élément'] == 664 ]

# Et je crée un DataFrame df_disponibilité contenant la disponibilité alimentaire + la population pour chaque pays (que je nettoie un peu)

df_disponibilité = pd.merge(df_food_supply, df_population, on='Code Pays', how='left')
df_disponibilité.drop(['Code Année', 'Pays_y', 'Symbole', 'Description du Symbole'], axis = 1, inplace = True) 
```

```python
# Je crée une nouvelle colonne nommée "Dispo Kcal/Jour" qui est la multiplication de la disponibilité alimentaire 
# et de la population

df_disponibilité['Dispo Kcal/Jour'] = (df_disponibilité['Valeur_x'] * df_disponibilité['Valeur_y'] * 1000)
df_disponibilité
```

<B> Calculez pour chaque produit le ratio "énergie/poids", que vous donnerez en kcal/kg </B>

```python
# Il me faut donc le df_food_supply créé plus haut que je vais joindre à un DataFrame ne contenant que la 
# dispo alimentaire en quantité et créer une nouvelle colonne avec le ratio energie/poids

df_dispo_poids = df_bilan[ df_bilan['Code Élément'] == 645 ]
df_ratio_energie_poids = pd.merge(df_food_supply, df_dispo_poids, on=['Code Pays', 'Code Produit'], how='left')
df_ratio_energie_poids.drop(['Code Élément_x','Code Élément_y','Code Année_x', 'Pays_y','Code Élément_x','Symbole_y','Produit_y','Code Année_y','Année_y', 'Description du Symbole_y','Description du Symbole_x','Symbole_x'], axis = 1, inplace = True)
df_ratio_energie_poids['Ratio Energie/Poids'] = (df_ratio_energie_poids['Valeur_x'] / df_ratio_energie_poids['Valeur_y'])

# Je retire les lignes qui ont seulement la colonne ratio energie/poids = à NaN
df_ratio_energie_poids = df_ratio_energie_poids.dropna(subset = ['Ratio Energie/Poids'])

# et je retire les lignes où celui ci est égal à 0
df_ratio_energie_poids = df_ratio_energie_poids[df_ratio_energie_poids['Ratio Energie/Poids'] != 0]

df_ratio_energie_poids
```

<B> Calculez, pour les produits végétaux uniquement, la disponibilité intérieure mondiale exprimée en kcal </B>

```python
#Pour ça il me faut un nouveau DF concernant les groupes de produits pris sur le site de la FAO (Sans l'agregation Chine)

df_groupe_produits = pd.read_csv("FAOSTAT_Groupe de produits.csv")
df_groupe_produits.drop(['Code Domaine','Domaine','Code Année','Année','Symbole','Description du Symbole'], axis = 1, inplace = True)

#Je crée un DF ne contenant que la disponibilité alimentaire et les végétaux :

df_végétaux = df_groupe_produits[(df_groupe_produits['Code Élément'] == 664 ) & ( df_groupe_produits['Code Produit'] == 2903) ]

# Et je crée un DataFrame df_disponibilité_végétaux contenant la disponibilité alimentaire des végétaux + la population pour chaque pays

df_disponibilité_végétaux = pd.merge(df_végétaux, df_population, on='Code Pays', how='left')

#Je crée une nouvelle colonne Dispo totale/jour

df_disponibilité_végétaux['Dispo totale/jour'] = (df_disponibilité_végétaux['Valeur_x'] * df_disponibilité_végétaux['Valeur_y'] * 1000)

# J'agrége toutes ces données pour avoir le total mondial en Kcal
print(df_disponibilité_végétaux["Dispo totale/jour"].sum(), "Kcal")


```

<B> Question : en termes de disponibilité intérieure de produits végétaux, combien d'humains pourraient être nourris si toute la production végétale était utilisée pour de la nourriture ? </B>

```python
#Recommandation FAO de 2414 Kcal/jour (je divise le résultats pour avoir en milliards)

print(((df_disponibilité_végétaux["Dispo totale/jour"].sum()/2414)/100000000),"milliards de personnes")
```
